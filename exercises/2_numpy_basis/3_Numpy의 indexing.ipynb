## 문제 1. np.concatenate 후의 shape 예측하기
* 위의 중간고사, 기말고사 성적표 데이터를 axis`1`을 기준으로 concatenate 했을 때의 shape를 구해주세요

````python
# 우선 실행해주세요
middle_scores = np.array([
    [80,92,70,65,92],
    [91,75,90,68,85],
    [86,76,42,72,88],
    [77,92,52,60,80],
    [75,85,85,92,95],
    [96,90,95,81,72]
])

final_scores = np.array([
    [85,95,90,66,93],
    [93,70,80,60,81],
    [89,78,55,75,80],
    [80,94,59,72,90],
    [70,82,81,95,72],
    [90,76,93,82,89]
])
````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과는 `answer1`에 담아주세요

````python
answer1 = (6,10)
````

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

````python
assert 'answer1' in globals(), "결과를 answer1이라는 변수에 담아주세요!, answer1의 값을 보고 평가합니다."

if (answer1 == np.concatenate((middle_scores, final_scores), axis=1).shape):
    print("정답입니다!!!!")
    print("concatenate한 결과가 무엇을 의미하는지 잘 생각해보세요!")
else:
    raise ValueError("shape가 맞지 않습니다. 다시 생각해보세요!")
````

## 문제 5. 행렬의 shape 변경하기(2)

* 수능을 대비하여 학생들에게 3개월마다의 성적의 평균을 저장하고 있는 표를 만들어주세요

정답 >>>
```python

np.array([[388.33333333, 376.33333333, 373.66666667],
       [302.66666667, 332.        , 353.        ],
       [181.33333333, 166.        , 169.66666667],
       [273.        , 303.33333333, 385.33333333],
       [188.        , 182.66666667, 186.66666667]])
```

* 참고 : numpy에서의 평균

numpy에서 축별로 평균을 구하는 방법은 다음과 같습니다.

`numpy_array.mean(axis=n)`

예시 >
```python
A = np.array([1,3,5,7,9])
A.mean(axis=0) # 결과 >> 5.0
```

````python
# 우선 실행해주세요
scores = np.array([
    [390,382,393,362,375,392,358,367,396],
    [300,297,311,320,336,340,342,357,360],
    [187,190,167,164,163,171,178,151,180],
    [298,267,254,302,300,308,380,384,392],
    [96,78,390,80,72,396,68,92,400]
])
````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과는 `answer5`에 담아주세요

````python
sorted_scores = scores.reshape(5,3,3)
answer5 = sorted_scores.mean(axis=2)
````

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

````python
from numpy.testing import assert_array_almost_equal

assert 'answer5' in globals(), "결과를 answer5라는 변수에 담아주세요!, answer5의 값을 보고 평가합니다."

A = np.array([[388.33333333, 376.33333333, 373.66666667],
       [302.66666667, 332.        , 353.        ],
       [181.33333333, 166.        , 169.66666667],
       [273.        , 303.33333333, 385.33333333],
       [188.        , 182.66666667, 186.66666667]])

assert_array_almost_equal(A,answer5,decimal=2,
                          err_msg="\n\n정답은 \n{}\n입니다. 코드를 다시 확인해주세요".format(A))

print("정답입니다!!!")
````
